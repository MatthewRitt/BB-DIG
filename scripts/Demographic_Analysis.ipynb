Alabama Voting Demographics

```python
import pandas as pd

target_counties = [
    'SUMTER', 'CHOCTAW', 'GREENE', 'HALE', 'MARENGO', 'PERRY', 
    'DALLAS', 'WILCOX', 'LOWNDES', 'BUTLER', 'CRENSHAW', 
    'MONTGOMERY', 'PIKE', 'BULLOCK', 'MACON', 'BARBOUR', 'RUSSELL'
]

final_columns = [
    'County',
    'Eligible Black Voters', 'Black Voters (Turnout)', 'Black: Eligible but Did Not Vote',
    'Eligible White Voters', 'White Voters (Turnout)', 'White: Eligible but Did Not Vote',
    'Eligible Hispanic Voters', 'Hispanic Voters (Turnout)', 'Hispanic: Eligible but Did Not Vote',
    'Eligible Asian Voters', 'Asian Voters (Turnout)', 'Asian: Eligible but Did Not Vote'
]

# 2024
l2_24 = pd.read_csv('AL_l2_2024_gen_stats_2020county.csv')
cvap_24 = pd.read_csv('al_cvap_2024_cnty.csv')

l2_24['countyname'] = l2_24['countyname'].str.upper().str.strip()
t24 = l2_24[l2_24['countyname'].isin(target_counties)].copy()
t24.rename(columns={'countyname': 'County', 'voted_eur': 'White Voters (Turnout)', 'voted_aa': 'Black Voters (Turnout)', 'voted_hisp': 'Hispanic Voters (Turnout)', 'voted_esa': 'Asian Voters (Turnout)'}, inplace=True)

cvap_24['County'] = cvap_24['COUNTY'].str.replace(' County', '', regex=False).str.upper().str.strip()
c24 = cvap_24[['County', 'CVAP_WHT24', 'CVAP_BLA24', 'CVAP_HSP24', 'CVAP_ASI24']].copy()
c24.rename(columns={'CVAP_WHT24': 'Eligible White Voters', 'CVAP_BLA24': 'Eligible Black Voters', 'CVAP_HSP24': 'Eligible Hispanic Voters', 'CVAP_ASI24': 'Eligible Asian Voters'}, inplace=True)

df_2024 = pd.merge(t24[['County', 'White Voters (Turnout)', 'Black Voters (Turnout)', 'Hispanic Voters (Turnout)', 'Asian Voters (Turnout)']], c24, on='County', how='inner')

df_2024['White: Eligible but Did Not Vote'] = (df_2024['Eligible White Voters'] - df_2024['White Voters (Turnout)']).apply(lambda x: max(0, x))
df_2024['Black: Eligible but Did Not Vote'] = (df_2024['Eligible Black Voters'] - df_2024['Black Voters (Turnout)']).apply(lambda x: max(0, x))
df_2024['Hispanic: Eligible but Did Not Vote'] = (df_2024['Eligible Hispanic Voters'] - df_2024['Hispanic Voters (Turnout)']).apply(lambda x: max(0, x))
df_2024['Asian: Eligible but Did Not Vote'] = (df_2024['Eligible Asian Voters'] - df_2024['Asian Voters (Turnout)']).apply(lambda x: max(0, x))

df_2024 = df_2024[final_columns]
df_2024.to_csv('2024_17_Counties_Demographics.csv', index=False)
print("2024 File Generated: 2024_17_Counties_Demographics.csv")


# 2020
l2_20_blocks = pd.read_csv('AL_l2_turnout_stats_block20.csv')
cvap_20 = pd.read_csv('al_cvap_2020_cnty.csv')

l2_20_blocks['COUNTYFP'] = l2_20_blocks['geoid20'].astype(str).str.zfill(15).str[2:5].astype(int)
fips_to_county = {5: 'BARBOUR', 11: 'BULLOCK', 13: 'BUTLER', 23: 'CHOCTAW', 41: 'CRENSHAW', 47: 'DALLAS', 63: 'GREENE', 65: 'HALE', 85: 'LOWNDES', 87: 'MACON', 91: 'MARENGO', 101: 'MONTGOMERY', 105: 'PERRY', 109: 'PIKE', 113: 'RUSSELL', 119: 'SUMTER', 131: 'WILCOX'}
l2_20_blocks['County'] = l2_20_blocks['COUNTYFP'].map(fips_to_county)
l2_20_target_blocks = l2_20_blocks.dropna(subset=['County'])

t20 = l2_20_target_blocks.groupby('County')[['g20201103_voted_eur', 'g20201103_voted_aa', 'g20201103_voted_hisp', 'g20201103_voted_esa']].sum().reset_index()
t20.rename(columns={'g20201103_voted_eur': 'White Voters (Turnout)', 'g20201103_voted_aa': 'Black Voters (Turnout)', 'g20201103_voted_hisp': 'Hispanic Voters (Turnout)', 'g20201103_voted_esa': 'Asian Voters (Turnout)'}, inplace=True)

cvap_20['County'] = cvap_20['COUNTY'].str.replace(' County', '', regex=False).str.upper().str.strip()
c20 = cvap_20[['County', 'CVAP_WHT20', 'CVAP_BLK20', 'CVAP_HSP20', 'CVAP_ASN20']].copy()
c20.rename(columns={'CVAP_WHT20': 'Eligible White Voters', 'CVAP_BLK20': 'Eligible Black Voters', 'CVAP_HSP20': 'Eligible Hispanic Voters', 'CVAP_ASN20': 'Eligible Asian Voters'}, inplace=True)

df_2020 = pd.merge(t20, c20, on='County', how='inner')

df_2020['White: Eligible but Did Not Vote'] = (df_2020['Eligible White Voters'] - df_2020['White Voters (Turnout)']).apply(lambda x: max(0, x))
df_2020['Black: Eligible but Did Not Vote'] = (df_2020['Eligible Black Voters'] - df_2020['Black Voters (Turnout)']).apply(lambda x: max(0, x))
df_2020['Hispanic: Eligible but Did Not Vote'] = (df_2020['Eligible Hispanic Voters'] - df_2020['Hispanic Voters (Turnout)']).apply(lambda x: max(0, x))
df_2020['Asian: Eligible but Did Not Vote'] = (df_2020['Eligible Asian Voters'] - df_2020['Asian Voters (Turnout)']).apply(lambda x: max(0, x))

df_2020 = df_2020[final_columns]
df_2020.to_csv('2020_17_Counties_Demographics.csv', index=False)
print("2020 File Generated: 2020_17_Counties_Demographics.csv")
```
