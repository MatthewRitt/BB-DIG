```python
import pandas as pd

df_2024 = pd.read_csv('2024_17_Counties_Demographics.csv')


al_02_counties = [
    'BARBOUR', 'BULLOCK', 'BUTLER', 'CRENSHAW', 
    'MACON', 'MONTGOMERY', 'PIKE', 'RUSSELL'
]

district_2_data = df_2024[df_2024['County'].isin(al_02_counties)].copy()


columns_to_keep = [
    'County',
    'Eligible Black Voters', 'Black Voters (Turnout)', 'Black: Eligible but Did Not Vote',
    'Eligible White Voters', 'White Voters (Turnout)', 'White: Eligible but Did Not Vote',
    'Eligible Hispanic Voters', 'Hispanic Voters (Turnout)', 'Hispanic: Eligible but Did Not Vote'
]
district_2_data = district_2_data[columns_to_keep]

district_2_data.rename(columns={
    'Eligible Black Voters': 'African American: Eligible',
    'Black Voters (Turnout)': 'African American: Voted',
    'Black: Eligible but Did Not Vote': 'African American: Missing Voters',
    
    'Eligible White Voters': 'White: Eligible',
    'White Voters (Turnout)': 'White: Voted',
    'White: Eligible but Did Not Vote': 'White: Missing Voters',
    
    'Eligible Hispanic Voters': 'LatinX: Eligible',
    'Hispanic Voters (Turnout)': 'LatinX: Voted',
    'Hispanic: Eligible but Did Not Vote': 'LatinX: Missing Voters'
}, inplace=True)

district_2_data.to_csv('AL_District_2_Voter_Gap_2024.csv', index=False)

district_2_data
```
